Task:

Extract only columns 'W' and 'Z' from the DataFrame and Compute the sum of each row and add it as a new column 'Sum' to the DataFrame.


```python
import pandas as pd

# Creating the DataFrame
df = pd.DataFrame({
    'W': [2.7, 0.7, -2.0, 0.2, 0.2],
    'X': [0.6, -0.3, 0.7, -0.8, 2.0],
    'Y': [0.9, -0.8, 0.5, -0.9, 2.6],
    'Z': [0.5, 0.6, -0.5, 1.0, 0.7]
}, index=['A', 'B', 'C', 'D', 'E'])

# Extracting columns 'W' and 'Z'
df_selected = df[['W', 'Z']]

# Computing the sum of each row and adding it as a new column 'Sum'
df_selected['Sum'] = df_selected.sum(axis=1)

# Display the updated DataFrame
print(df_selected)
```

```python
import pandas as pd

# Creating the DataFrame
df = pd.DataFrame({
    'W': [2.7, 0.7, -2.0, 0.2, 0.2],
    'X': [0.6, -0.3, 0.7, -0.8, 2.0],
    'Y': [0.9, -0.8, 0.5, -0.9, 2.6],
    'Z': [0.5, 0.6, -0.5, 1.0, 0.7]
}, index=['A', 'B', 'C', 'D', 'E'])

# Extracting columns 'W' and 'Z' and creating a copy
df_selected = df[['W', 'Z']].copy()

# Computing the sum of each row and adding it as a new column 'Sum'
df_selected['Sum'] = df_selected.sum(axis=1)

# Display the updated DataFrame
print(df_selected)
```
