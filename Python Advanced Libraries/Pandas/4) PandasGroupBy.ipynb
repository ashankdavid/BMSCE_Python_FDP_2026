# Groupby

<b>Groupby</b> Allows you to group together rows, based off of a column and perform a aggregate function on them

```python
import numpy as np
import pandas as pd
```

```python
data ={'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Ashank', 'Saurav', 'Chandan', 'Nikhil', 'Tanmay', 'Ujjwal'], 
       'Sales':[200,120,340,124,243,350]} #Creating Dict of data
```

```python
df = pd.DataFrame(data) #Creating dataframe out of dict
```

```python
df
```

```python
df.groupby('Company') # This will return a groupby object
```

```python
byComp = df.groupby('Company') # Stored the object and accessed aggregate func()
```

```python
byComp.mean() # Aggregate Function mean()
```

```python
byComp.sum() # Aggregate Function sum()
```

```python
byComp.std() # Aggregate Function std() (Standard Deviation)
```

```python
byComp.sum().loc['FB'] #Calc the sum of only FB
```

```python
df.groupby('Company').sum().loc['FB'] # Doing same in one line
```

```python
df.groupby('Company').count() # Aggregate Function count()
```

```python
df.groupby('Company').max() # Aggregate Function max()
```

```python
df.groupby('Company').min() # Aggregate Function min()
```

```python
df.groupby('Company').describe() # Aggregate Function describe() (Auto)
```

```python
df.groupby('Company').describe().transpose() # Aggregate Function transpose()
```

```python
df.groupby('Company').describe().transpose()['FB'] # for FB
```
