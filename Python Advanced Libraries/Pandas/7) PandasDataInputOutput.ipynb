```python
df = pd.read_csv('example')
```

```python
df.to_csv('My_output')
```

```python
pd.read_csv('My_output')
```

In the above example, the index gets saved as a new unnamed column!, so to prevent that we need to say a new argument as, "index=False" which wil not save the index as a new unnamed column!

```python
df.to_csv('My_output', index=False)
```

```python
pd.read_csv('My_output')
```

In the above example, we said index = False, so that the index doesn't get saved as a new column!
