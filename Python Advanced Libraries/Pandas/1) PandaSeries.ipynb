# Series

A Series is very similar to a NumPy array, infact it is built on a NumPy array object, but what makes it different from a NumPy array is Series can have labels, meaning they can be indexed by a label, lets understand it:

```python
import numpy as np
```

```python
import pandas as pd
```

```python
labels = ['a','b','c']
my_data = [10,20,30]
arr = np.array(my_data)
d = {'a':10, 'b':20, 'c':30}
```

```python
labels # Just printing the labels
```

```python
my_data # Just printing my_data
```

```python
arr # Just printing the arr
```

```python
d # just printing the Dictionary
```

```python
pd.Series(data = my_data) # Creating a Series and passing my_data in it
```

```python
pd.Series(data = my_data, index = labels) # Passing both my_data and labels
```

```python
pd.Series(my_data, labels) # Directly passing my_data and Labels into Series
```

```python
pd.Series(arr) # You can also pass Arrays in the Series
```

```python
pd.Series(arr, labels) # Can pass arrays with labels also in the Series
```

```python
pd.Series(d) # passing Dictionary will automatically set indexes for data
```

```python
d # Well Dict was looking like this!
```

***
A series can also hold many other types of Data in its Data Points not only int but also characters

```python
pd.Series(data = labels) # Passing Characters as data in Series
```

```python
pd.Series(["Coders",  "Arcade"]) # Passing Strings also is Possible
```

```python
pd.Series(["Coders",  "Arcade"], ["C", "A"]) # Passing Chars in Indexes also
```

***
You can also hold inbuilt python functions also in Pandas Series

```python
pd.Series(data = [sum,print,len])
```

Well you wont do this ever but still, this tells us that you can almost hold anything in a Pandas Series
