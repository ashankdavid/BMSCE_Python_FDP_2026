```python
import numpy as np
import pandas as pd
```

# Finding unique values in DataFrame

<b> .unique() method </b>

```python
df['col2'].unique()
```

<b> .nunique method for Length </b>

```python
len(df['col2'].unique()) # Normally finding Length
```

```python
df['col2'].nunique() # Finding Length using nunique
```

<b> counting unique data </b>

```python
df['col2'].value_counts()
```

# Selecting Data

```python
df
```

<b> Conditional Selction </b>

```python
df[df['col2']>2]
```

```python
df[(df['col1']>2) & (df['col2']==444)]
```

<b> Apply Method </b><br>
We can use our custom functions to work with a DataFrame using apply method

```python
def times2(x):  # Imagine we have a function
    return x*2  
```

```python
df['col1'].sum() # Normal sum inbuilt function can be applied like this
```

```python
df['col1'].apply(times2) # we can use apply method for a custom function we made
```

```python
df['col3'] # printing col3
```

```python
df['col3'].apply(len) # We can apply inbuilt functions also
```

```python
df['col1'].apply(lambda x: x*2)
```

If you are using a funciton once only u can pass a lambda function in apply method so there only the function will be used and this will save out time! and you also dont have to search for a function in order to apply it!

<b> Removing Columns </b>

# Pandas Pivot Table

```python
data = {'A':['foo','foo','foo','bar','bar','bar'],
       'B':['one','one','one','two','two','one'],
       'C':['x','y','x','y','x','y'],
       'D':[1,3,2,5,4,1]}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.pivot_table(values='D',index=['A','B'],columns=['C'])
```
