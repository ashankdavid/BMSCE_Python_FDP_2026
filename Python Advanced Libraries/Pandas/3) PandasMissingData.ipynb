# Handling Missing Data in Pandas

```python
import numpy as np
import pandas as pd
```

```python
d = {'A':[1,2,np.nan], 'B':[5,np.nan,np.nan], 'C':[1,2,3]} # Creating Dict!
```

```python
df = pd.DataFrame(d) # Creating DataFrame out of that Dict!
```

```python
df # Printing the df
```

# Dropping Missing Data

```python
df.dropna() # Using dropna() we can drop the rows by default with nan values
```

```python
df.dropna(axis=1) # To drop the columns change the axis to 1!
```

Now we can specify a Threshold also, to control the dropping of data!

Example:

```python
df # printing the df 
```

```python
df.dropna(thresh=2) #this will not allow to drop the row which contained 2 value
```

# Filling Missing Data

```python
df
```

```python
df.fillna(value='Fill Value') # This will fill the nan places with Values Arg!
```

Now, Lets say you wanna fill the mean of the data in some particular na value

```python
df['A'] # Printing A from df
```

```python
df['A'].fillna(value=df['A'].mean()) # Filling mean of the data using fillna
```
