# Matplotlib - Part 1

```python
import matplotlib.pyplot as plt # importing matplotlib
```

```python
%matplotlib inline
```

```python
# Creating Data
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

<b>There are 2 ways to create matplotlib plots</b> <br>
1) Functional Way <br>
2) Object Oriented Way
***
1) Functional Way

```python
plt.plot(x,y) # ploting using matplotlib
```

```python
plt.plot(x,y,'r') # Adding color
```

```python
plt.plot(x,y) # plotting
plt.xlabel('X Label') # putting label for x - axis
plt.ylabel('Y Label') # Putting label for y - axis
plt.title('Title') # Putting Plot Title
```

<b> Multi-Plot on same canvas</b>

```python
# First Plot
plt.subplot(1,2,1) #(row, column, plotNo)
plt.plot(x,y,'r') #(x, y, color(red))

#Second Plot
plt.subplot(1,2,2) #(row, column, plotNo)
plt.plot(y,x,'b') #(x, y, color(blue))
```

***
2) Object Oriented Way

```python
fig = plt.figure() # Step 1- Created a Figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #(x-axis, y-axis, width, height) of axes (2)
axes.plot(x,y) # (step 3)
axes.set_xlabel('X Label') # setting x label
axes.set_ylabel('Y Label') # setting y label
axes.set_title('Title') # setting title
```

Now, we can add sub axes also inside some other axes:

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_xlabel('X label Out')
axes1.set_ylabel('Y label Out')
axes1.set_title('Larger Plot')
axes2.set_xlabel('X label In')
axes2.set_ylabel('Y label In')
axes2.set_title('Smaller Plot')

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
# More Examples:
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.45,0.25,0.4,0.3])
axes1.set_xlabel('X label Out')
axes1.set_ylabel('Y label Out')
axes1.set_title('Larger Plot')
axes2.set_xlabel('X label In')
axes2.set_ylabel('Y label In')
axes2.set_title('Smaller Plot')

# Plotting
axes1.plot(np.arange(8).reshape(2,4),np.arange(8).reshape(2,4))
axes2.plot(np.arange(8).reshape(2,4),np.arange(8).reshape(2,4))
```

```python
# Revision OOPs
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

# Matplotlib - Part 2

Plotting Sub Plots using Matplotlib

In this we use "plt.subplots()" to plot subplots in matplotlib, here "plt.subplots()" is doing the same job of "fig.add_axes()", based on how many rows and columns you want

so we write "fig,axes = plt.subplot()"

```python
fig,axes = plt.subplots()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout # Fixing overlaps
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes
```

<b>In the above cell we printed axes and we found that axes is just an array of axes for different plots, that means we can iterate over it and we can do plotting from there!</b>

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
for current_axes in axes:
    current_axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[0].set_xlabel("X First Label")
axes[0].set_ylabel("Y First Label")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")
axes[1].set_xlabel("X Second Label")
axes[1].set_ylabel("Y Second Label")

plt.tight_layout()
```

<b>Adding Legends to a figure:</b>

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# Plotting
ax.plot(x,x**2, label = 'X Sqaure') # adding legend label for Square here
ax.plot(x,x**3, label = 'X Cube') # adding legend label for Cube here

# Adding Legends
ax.legend(loc=10)

# setting title and labels
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

### Control over axis appearance

In this section we will look at controlling axis sizing properties in matplotlib figure

# Plot Range

We can configure the ranges of the axes using set_ylim and set_xlim method in the axis object, or axis('tight') for automatically getting "tightly fitted" axes ranges: 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='red',lw='2', ls='--')

ax.set_xlim([0,1])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='red',lw='2', ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

# Special Plot Types

There are many specialized plots we can create, such as barplots, histograms, scatter plots, and much more. Most of these type of plots we will actually create using seaborn, a statistical plotting library for Python. But here are a few examples of these type of plots:

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```
