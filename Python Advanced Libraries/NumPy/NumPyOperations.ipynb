# NumPy Operations

• <b>Array with Array</b> <br>
• <b>Array with Scalars</b> <br>
• <b>Universal array Functions</b>

```python
import numpy as np
```

```python
arr = np.arange(0,11)
```

```python
arr
```

# Array with Array

```python
arr + arr # Adding 2 arrays
```

```python
arr - arr # Subtracting 2 array
```

```python
arr * arr
```

# Array with Scalar

```python
arr + 100 # Scalar Add
```

```python
arr * 100 # Scalar Multiply
```

```python
arr - 100
```

```python
arr ** 2 # Power of 2
```

# Error vs Warning in NumPy

```python
1/0 # Error
```

```python
arr / arr #Warning (nan)
```

```python
1 / arr # Warning (Infinity)
```

# Universal Array Functions

Numpy comes with many Universal Array Functions prebuilt, which you can use to perform different mathematical operations

```python
np.sqrt(arr) # Square Root Function
```

```python
np.exp(arr) # Exponent Function
```

```python
np.max(arr) # Maximum Function for Arrays
```

```python
arr.max() # Directly using max() Function on Arrays
```

```python
np.sin(arr) # Sin Trigonometric function
```
